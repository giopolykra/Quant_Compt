```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram
```

```python
n=8
n_q = 8
n_b = 8
qc_output = QuantumCircuit(n_q,n_b)
for j in range(n):
    qc_output.measure(j,j)
qc_output.draw(output='mpl')
```

```python
counts = execute(qc_output,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

```python
qc_encode = QuantumCircuit(n)
qc_encode.x(7)
qc_encode.draw(output='mpl')
```

```python
qc = qc_encode + qc_output
qc.draw(output='mpl')
```

```python
counts = execute(qc, Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

```python
qc_encode = QuantumCircuit(n)
qc_encode.x(1)
qc_encode.x(5)
qc_encode.draw(output='mpl')
```

```python
qc_cnot = QuantumCircuit(2,2)
qc_cnot.barrier()
qc_cnot.cx(0,1)
qc_cnot.barrier()
qc_cnot.measure(0,0)
qc_cnot.measure(1,1)
qc_cnot.draw(output='mpl')
```

```python
counts = execute(qc_cnot,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

```python
counts = execute(qc_cnot,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

```python
qc_xx = QuantumCircuit(4,2)
qc_xx.x(0)
qc_xx.x(1)
qc_xx.barrier()
qc_xx.cx(0,2)
qc_xx.cx(1,3)   
qc_xx.ccx(0,1,2)
qc_xx.barrier()
qc_xx.measure(2,0)
qc_xx.measure(3,1)
qc_xx.draw(output='mpl')
```

```python
counts = execute(qc_xx,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

```python
qc = QuantumCircuit(2,2)
qc.x(0)
qc.barrier()
qc.cx(0,1)
qc.barrier()
qc.measure(0,0)
qc.measure(1,1)
qc.draw(output='mpl')
```

```python
counts = execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

# The Unique Properties of Qubits

```python
from qiskit import *
from qiskit.visualization import plot_histogram
from matplotlib import pyplot as plt


```

```python
measure_z = QuantumCircuit(1,1)
measure_z.measure(0,0)
measure_z.draw(output='mpl')
```

```python
measure_x = QuantumCircuit(1,1)
measure_x.h(0)
measure_x.measure(0,0)
measure_x.draw(output='mpl')
```

```python
qc_0 = QuantumCircuit(1)
qc_0.draw(output= 'mpl')
```

```python
qc = qc_0 + measure_z
print('Results for z measurement:')
counts = execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```
