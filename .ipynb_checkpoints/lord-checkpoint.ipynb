# Bell State

```python
from matplotlib import pyplot as plt
import numpy as np
from qiskit import *
```

```python
def bell_state(Nqubits,param):
    qr = QuantumRegister(Nqubits)
    cr = ClassicalRegister(Nqubits)
    qc = QuantumCircuit(qr, cr)
    for j in range(Nqubits):
        if param[j] == 1:
            qc.x(qr[j])
    qc.h(qr[0])
    qc.cx(qr[0],qr[1])
    [qc.measure(qr[i], cr[i]) for i in range(Nqubits)]
    return qc
```

```python
#The param controls the input state by adding an x gate to the corresponding qubit. 
#So by we can change the input state from |00> to |01> by changing param=[0,0] to param=[0,1].
#In this way we can later simulate all the different bell states
```

```python
import numpy as np
param = np.random.randint(2, size=2) #Create a random input state just for fun
qc = bell_state(2,param)
print('The random import state is |',param[0], param[1],'>')
qc.draw(output = 'mpl')
```

```python
#Simulation
```

```python
from qiskit.visualization import plot_histogram
print('Results for measurement:')
counts = execute(bell_state(2,param),Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

# GHZ state

```python
#Creation of a GHZ state from an |000...0> input state
```

```python
def ghz_state(Nqubits):
    qr = QuantumRegister(Nqubits)
    cr = ClassicalRegister(Nqubits)
    qc = QuantumCircuit(qr, cr)
    qc.h(qr[0])
    [qc.cx(qr[j],qr[j+1]) for j in range(Nqubits-1)]
    qc.barrier()
    [qc.measure(qr[i], cr[i]) for i in range(Nqubits)]
    return qc
```

```python
N = 5
qc = ghz_state(N)
print('The GHZ state for N =',N,'qubits' )
qc.draw(output = 'mpl')
```

```python
from qiskit.visualization import plot_histogram
print('Results for measurement:')
counts = execute(ghz_state(N),Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```

## Adding parameters for optimization

```python
def ghz_var(Nqubits,param):
    qr = QuantumRegister(Nqubits)
    cr = ClassicalRegister(Nqubits)
    qc = QuantumCircuit(qr, cr)
    qc.h(qr[0])
#    qc.rx(param[0],qr[1])
#    qc.rz(param[1],qr[1])
#    qc.rx(param[0],qr[1])
    for j in range(Nqubits-1):
        qc.cx(qr[j],qr[j+1])
    qc.barrier()
    [qc.measure(qr[i], cr[i]) for i in range(Nqubits)]
    return qc
```

```python
N = 3
param_test = np.random.rand(2)
qc = ghz_var(N,param_test)
qc.draw(output='mpl')
```

```python
from qiskit.visualization import plot_histogram
print('Results for measurement:')
counts = execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
print('The two randomly selectet parameters:',param_test)
plot_histogram(counts)
```

```python
import numpy as np
np.random.seed(999999)
target_distr = np.random.rand(2)
#we define 4 parameters because we know we are going to end up with 4 states after measurement
# We now convert the random vector into a valid probability vector
target_distr /= sum(target_distr)
print(target_distr ,sum(target_distr))
```

```python
from qiskit import Aer, execute
#Backend represent either a simulator or a real quantum computer
backend = Aer.get_backend("qasm_simulator") #runs the quantum circuit
NUM_SHOTS = 10000

def get_probability_distribution(counts):
    output_distr = [v / NUM_SHOTS for v in counts.values()]
    if len(output_distr) == 1:
        output_distr.append(0)
    return output_distr

def obj_function(param):
    # Obtain the quantum circuit from the random paramters
    qc = ghz_var(N,param)
    # Execute the quantum circuit to obtain the probability distribution associated with the current parameters
    result = execute(qc, backend, shots=NUM_SHOTS).result()
    # Obtain the counts for each measured state, and convert those counts into a probability vector
    output_distr = get_probability_distribution(result.get_counts(qc))
    # Calculate the cost as the distance between the output distribution and the target distribution
    cost = sum([np.abs(output_distr[i] - target_distr[i]) for i in range(2)])
    return cost
```

```python
from qiskit.aqua.components.optimizers import COBYLA
#this is a numerical optimization method for constrained problems 
#where the derivative of the objective function (in this case 'cost') is not known
# Initialize the COBYLA optimizer
optimizer = COBYLA(maxiter=5, tol=1)

ret = optimizer.optimize(num_vars=2, objective_function=obj_function, initial_point=param_test)
#the definition: optimize(num_vars, objective_function, gradient_function=None, variable_bounds=None, initial_point=None)

# Obtain the output distribution using the final parameters
qc = ghz_var(N,ret[0])
counts = execute(qc, backend, shots=NUM_SHOTS).result().get_counts(qc)
output_distr = get_probability_distribution(counts)

print(counts,"\n")
print(get_probability_distribution(counts),"\n")
print("Target Distribution:", target_distr,"\n")
print("Obtained Distribution:", output_distr,"\n")
print("Output Error (Manhattan Distance):", ret[1],"\n")
print("Parameters Found:", ret[0],"\n")
#print(ret)
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2], target_distr, 'r+',markeredgewidth=3,markersize=15,label="target")
plt.plot([1,2], output_distr, 'b+',markeredgewidth=3,markersize=15,label="obtained")
plt.xlabel('Parameter #')
plt.ylabel('Value')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.suptitle('NUM_SHOTS = 10k');
```

# 3-qubit Ising model

$H = -\sum_{<1,j>}J_{ij}\sigma_{1}\sigma_{j}$ where each spin can be in the $+1$ or $-1$ spin state and $J_{ij}$ are the nearest-neighbour coupling strengths. We choose the coumpling matrix $J=\begin{pmatrix}
0 &  1 &  0 \\
1 &  0 & -1 \\
0 & -1 &  0
\end{pmatrix}$ which we can simply represent as $J=[J_{1}, J_{2}]=[1,-1]$

```python
from matplotlib import pyplot as plt
import numpy as np
from qiskit import *
def circuit(Nqubits,param1,param2):
    qr = QuantumRegister(Nqubits)
    cr = ClassicalRegister(Nqubits)
    qc = QuantumCircuit(qr, cr)
    qc.u3(param1[0],param1[1],param1[2],qr[0])
    qc.u3(param2[0],param2[1],param2[2],qr[1])
    for j in range(Nqubits):
        qc.h(qr[j])
    qc.barrier()
    [qc.measure(qr[i], cr[i]) for i in range(Nqubits-2)]
    return qc

#We want to return the expectation value of measurment for each qubit
```

```python
import numpy as np
param1 = [0,np.pi/4,0]
param2 = [0,np.pi,0]

qc = circuit(3,param1,param2)
qc.draw(output ='mpl')
```

```python
from qiskit.visualization import plot_histogram
print('Results for measurement:')
counts = execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
print(counts)
plot_histogram(counts)
```

# Quantum nodes

```python
def ansatz(var):
    qml.Rot(0.3, 1.8, 5.4, wires=1)
    qml.RX(var[0], wires=0)
    qml.RY(var[1], wires=1)
    qml.CNOT(wires=[0, 1])
```

```python
@qml.qnode(dev)
def circuit_X(var):
    ansatz(var)
    return qml.expval(qml.PauliX(1))
```

```python
@qml.qnode(dev)
def circuit_Y(var):
    ansatz(var)
    return qml.expval(qml.PauliY(1))
```

# Objective

```python
# The cost function to be optimized in VQE is simply a linear combination
# of the expectations, which defines the expectation of the Hamiltonian we
# are interested in. In our case, we square this cost function to provide
# a more interesting landscape with the same minima.

def cost(var):
    expX = circuit_X(var)
    expY = circuit_Y(var)
    return (0.1 * expX + 0.5 * expY) ** 2
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import MaxNLocator

fig = plt.figure(figsize=(6, 4))
ax = fig.gca(projection="3d")

X = np.linspace(-3.0, 3.0, 20)
Y = np.linspace(-3.0, 3.0, 20)
xx, yy = np.meshgrid(X, Y)
Z = np.array([[cost([x, y]) for x in X] for y in Y]).reshape(len(Y), len(X))
surf = ax.plot_surface(xx, yy, Z, cmap=cm.coolwarm, antialiased=False)

ax.set_xlabel("v1")
ax.set_ylabel("v2")
ax.zaxis.set_major_locator(MaxNLocator(nbins=5, prune="lower"))

plt.show()
```

# Optimization

```python
from pennylane.optimize import GradientDescentOptimizer
opt = GradientDescentOptimizer(0.5)

var = [0.3, 2.5]
var_gd = [var]
for it in range(20):
    var = opt.step(cost, var)
    var_gd.append(var)
    print("Cost after step {:2d}: {: .7f} | Variables: [{: .5f},{: .5f}]".format(it + 1, cost(var), var[0], var[1]))
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.gca(projection="3d")

X = np.linspace(-3, np.pi / 2, 20)
Y = np.linspace(-3, 3, 20)
xx, yy = np.meshgrid(X, Y)
Z = np.array([[cost([x, y]) for x in X] for y in Y]).reshape(len(Y), len(X))

surf = ax.plot_surface(xx, yy, Z, cmap=cm.coolwarm, antialiased=False)

path_z = [cost(var) + 1e-8 for var in var_gd]
path_x = [v[0] for v in var_gd]
path_y = [v[1] for v in var_gd]
ax.plot(path_x, path_y, path_z, c="green", marker=".", label="graddesc", zorder=10)

ax.set_xlabel("v1")
ax.set_ylabel("v2")
ax.zaxis.set_major_locator(MaxNLocator(nbins=5, prune="lower"))

plt.legend()
plt.show()
```

# 3-qubit Ising Model in PyTorch (based on: https://pennylane.ai/qml/app/tutorial_isingmodel_PyTorch.html)

```python
import torch
from torch.autograd import Variable
import pennylane as qml
from pennylane import numpy as np
```

```python
def cirqdef(Nqubits):
    dev = qml.device("default.qubit", wires=Nqubits)
    return dev

dev = cirqdef(3)

@qml.qnode(dev, interface="torch")
def circuit(p1, p2):
    # We use the general Rot(phi,theta,omega,wires) single-qubit operation
    qml.Rot(p1[0], p1[1], p1[2], wires=1)
    qml.Rot(p2[0], p2[1], p2[2], wires=2)
    return [qml.expval(qml.PauliZ(i)) for i in range(3)]
```

```python
def cost(var1, var2):
    # the circuit function returns a numpy array of Pauli-Z expectation values
    spins = circuit(var1, var2)

    # the expectation value of Pauli-Z is +1 for spin up and -1 for spin down
    energy = -(1 * spins[0] * spins[1]) - (-1 * spins[1] * spins[2])
    return energy
```

```python
test1 = torch.tensor([0, np.pi, 0])
test2 = torch.tensor([0, np.pi, 0])
print(circuit(test1,test2))
#cost_check = cost(test1, test2)
#print("Energy for [1, -1, -1] spin configuration:", cost_check)
```

```python
torch.manual_seed(56)
p1 = Variable((np.pi * torch.rand(3, dtype=torch.float64)), requires_grad=True)
p2 = Variable((np.pi * torch.rand(3, dtype=torch.float64)), requires_grad=True)

var_init = [p1, p2]
cost_init = cost(p1, p2)

print("Randomly initialized angles:", var_init,"\n")
print("Corresponding cost before optimization:", cost_init,"\n")
```

## Optimization

```python
opt = torch.optim.SGD(var_init, lr=0.1)

def closure():
    opt.zero_grad()
    loss = cost(p1, p2)
    loss.backward()
    return loss

var_pt = [var_init]
cost_pt = [cost_init]
x = [0]

for i in range(100):
    opt.step(closure)
    if (i + 1) % 5 == 0:
        x.append(i)
        p1n, p2n = opt.param_groups[0]["params"]
        costn = cost(p1n, p2n)
        var_pt.append([p1n, p2n])
        cost_pt.append(costn)

        # for clarity, the angles are printed as numpy arrays
        print("Energy after step {:2d}: {: .7f} | Angles: {}".format(i+1, costn, [p1n.detach().numpy(), p2n.detach().numpy()]),"\n")
```

Things to keep in mind when using Pytorch optimizers:

1. loss.backward() computes the gradient of the cost function with respect to all parameters with requires_grad=True.
2. opt.step() performs the parameter update based on this current gradient and the learning rate.
3. opt.zero_grad() sets all the gradients back to zero. It’s important to call this before loss.backward() to avoid the accumulation of gradients from multiple passes.

```python
p1_final, p2_final = opt.param_groups[0]["params"]
print("Optimized angles:", p1_final, p2_final)
print("Final cost after optimization:", cost(p1_final, p2_final))
```

```python
import matplotlib
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6, 4))
plt.plot(x, cost_pt, label = 'global minimum')
plt.xlabel("Optimization steps")
plt.ylabel("Cost / Energy")
plt.legend()
plt.show()
```

#### Local minimum

If the spins are initialized close to the local minimum of zero energy, the optimizer is likely to get stuck here and never find the global minimum at -2.

```python
torch.manual_seed(9)
p3 = Variable((np.pi*torch.rand(3, dtype = torch.float64)), requires_grad = True)
p4 = Variable((np.pi*torch.rand(3, dtype = torch.float64)), requires_grad = True)

var_init_loc = [p3, p4]
cost_init_loc = cost(p3, p4)

print("Corresponding cost before optimization:", cost_init_loc)
```

```python
opt = torch.optim.SGD(var_init_loc, lr = 0.1)

def closure():
    opt.zero_grad()
    loss = cost(p3, p4)
    loss.backward()
    return loss

var_pt_loc = [var_init_loc]
cost_pt_loc = [cost_init_loc]

for j in range(100):
    opt.step(closure)
    if (j + 1) % 5 == 0:
        p3n, p4n = opt.param_groups[0]['params']
        costn = cost(p3n, p4n)
        var_pt_loc.append([p3n, p4n])
        cost_pt_loc.append(costn)

        # for clarity, the angles are printed as numpy arrays
        print('Energy after step {:5d}: {: .7f} | Angles: {}'.format(j+1, costn, [p3n.detach().numpy(), p4n.detach().numpy()]),"\n")
```

```python
fig = plt.figure(figsize=(6, 4))
plt.plot(x, cost_pt_loc, 'r', label = 'local minimum')
plt.xlabel("Optimization steps")
plt.ylabel("Cost / Energy")
plt.legend()
plt.show()
```

```python
def cost(var1, var2):
    # the circuit function returns a numpy array of Pauli-Z expectation values
    spins = circuit(var1, var2)
    # the expectation value of Pauli-Z is +1 for spin up and -1 for spin down
    energy = -(1 * spins[0] * spins[1]) - (-1 * spins[1] * spins[2])
    return energy
```
