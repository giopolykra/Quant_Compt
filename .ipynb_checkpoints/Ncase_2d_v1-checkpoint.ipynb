## For the $2^{nd}$ term:
$$\sum^{N}_{i=1}<\psi|\sigma^{X}_{i}|\psi> 
=<\psi|\sigma^{X}\otimes \mathbb{1}\otimes\cdots\otimes\mathbb{1}|\psi>
+<\psi|\mathbb{1}\otimes\sigma^{X}\otimes\cdots\otimes\mathbb{1}|\psi>
+\cdots
+<\psi|\mathbb{1}\otimes\mathbb{1}\otimes\cdots\otimes\sigma^{X}|\psi>
$$ 

### We can perform a transformation to the original state $|\psi>$ using the matrix 
$S=S^{-1}=\frac{1}{\sqrt{2}}\begin{pmatrix}
1 & 1\\
1 & -1
\end{pmatrix} = H$ so as to move to a basis where our observable $\sigma^{X}$ is diagonal.

$$
\begin{align}
<\psi|(\sigma^{X}\otimes\mathbb{1}\otimes\cdots\otimes\mathbb{1})|\psi>
&=<\psi|(H\otimes\mathbb{1}\otimes\cdots\otimes\mathbb{1})(\sigma^{Z}\otimes\mathbb{1}\otimes\cdots\otimes\mathbb{1})(H\otimes\mathbb{1}\otimes\cdots\otimes\mathbb{1})|\psi>\\
&=<\psi'_{1}|(\sigma^{Z}\otimes\mathbb{1}\otimes\cdots\otimes\mathbb{1})|\psi'_{1}>\\
&=<\psi'_{1}|\sigma^{Z}_{1}|\psi'_{1}>
\end{align}
$$

For the case $|\psi>=a|00>+b|01>+c|10>+d|11>$

$|\psi'_{1}>=H\otimes\mathbb{1}|\psi>=\frac{1}{\sqrt{2}}\left[(a+c)|00>+(b+d)|01>+(a-c)|10>+(b-d)|11>\right]$

$|\psi'_{2}>=\mathbb{1}\otimes H|\psi>=\frac{1}{\sqrt{2}}\left[(a+b)|00>+(a-b)|01>+(c+d)|10>+(c-d)|11>\right]$

$\begin{align}
\sum^{2}_{i=1}<\psi|\sigma^{X}_{i}|\psi>
&=<\psi'_{1}|\sigma^{Z}\otimes\mathbb{1}|\psi'_{1}>+<\psi'_{2}|\mathbb{1}\otimes\sigma^{Z}|\psi'_{2}>\\
&=\frac{1}{2}\left[|a+c|^2+|b+d|^2+|a-c|^2+|b-d|^2\right]
+\frac{1}{2}\left[|a+b|^2+|a-b|^2+|c+d|^2+|c-d|^2\right]
\end{align}$

```python
def conv_qubitbook(counts):
    N = len(str(list(counts.items())[0][0]))
    #I did that so that I wont have to include another input to the function
    qubitbook = counts
    for i in qubit_list(N):
        if i not in qubitbook.keys():
            qubitbook[i] = 0
    return qubitbook
```

```python
def qubit_list(Nqubit):
    li = ['0', '1']
    combs = [''.join(comb) for comb in product(li, repeat = Nqubit)]
    return combs
print(qubit_list(3))
```

## Couplings

```python
def couplings(N):
    h = []
    J = [] 
    for i in range(N):
        h.append(random.choice(range(1,2,2)))
    for j in range(N-1):
        J.append(random.choice(range(1,2,2)))
    return h,J

print("h=",couplings(3)[0],"\nJ=",couplings(3)[1])
```

## $1^{st}$ circuit for computation of $H1 = \sum^{N-1}_{i=0}J_{i}<\psi|\sigma^{Z}_{i}\otimes\sigma^{Z}_{i+1}|\psi>$

```python
def circuit(N,param,layers):
    qr = QuantumRegister(N)
    cr = ClassicalRegister(N)
    qc = QuantumCircuit(qr, cr)
    for l in range(layers):
        for i in range(N):
            qc.rx(param[i],qr[i])
            qc.ry(param[i+1],qr[i])
            qc.rz(param[i+2],qr[i])
        [qc.cx(qr[i],qr[i+1]) for i in range(N-1)]
        qc.barrier()
    #qc.x(qr[1])
    #qc.x(qr[2])
    qc.unitary
    [qc.measure(qr[i], cr[i]) for i in range(N)]
    return qc
```

```python
N=3
param = param_trial(N)
qc = circuit(N,param,1)
qc.draw(output="mpl")
```

```python
counts = execute(qc,Aer.get_backend('qasm_simulator'),shots = 100).result().get_counts()
print("counts:",counts)
counts = conv_qubitbook(counts)
counts = dict(sorted(counts.items()))
print("new_counts:",counts)
print("qubit list:",qubit_list(N))
plot_histogram(conv_qubitbook(counts))
```

Now we are ready to compute the expression  $<\psi>$

```python
def compute_H1(N,NUM_SHOTS,param,J,layers):
    H1=0
    a=[]
    qc = circuit(N,param,layers)
    job = execute(qc,Aer.get_backend('qasm_simulator'),shots=NUM_SHOTS)
    result = job.result()
    counts = result.get_counts()
    counts = conv_qubitbook(counts) 
    #adds the non-measured states to the list to complete the set of possible 2**N configurations
    counts = dict(sorted(counts.items())) #sorts the dictionary of counts according to their key
    [a.append([int(d) for d in str(list(counts)[i])]) for i in range(len(list(counts)))]
    #print("len(a):",len(a))
    shots = sum(counts.values())
    P1 = [(i, c/shots) for i, c in counts.items()]
    #print("probs:",P1)
    for i in range(N-1): #runs over the N-1 pairs of consecutive qubits
        for j in range(2**N): #runs over the set of 2**N states
            if a[j][i]==a[j][i+1]:
                H1 += J[i]*P1[j][1]
            if a[j][i]!=a[j][i+1]:
                H1 -= J[i]*P1[j][1]
            #print("state:'{}' (qubits:{} and {}) H1={}".format(P[j][0],i,i+1,H1))
    #print("H1 =",H1)
    return H1
```

```python
NUM_SHOTS = 1000
N=3
J = couplings(N)[1]
compute_H1(N,NUM_SHOTS,param,J,2)
```

## $2^{nd}$ circuit for computation of $H2 = \sum^{N}_{i=1}h_{i}<\phi|\sigma^{X}_{i}|\phi>$

```python
def circuit2(N,param,layers):
    qr = QuantumRegister(N)
    cr = ClassicalRegister(N)
    qc = QuantumCircuit(qr, cr)
    for l in range(layers):
        for i in range(N):
            qc.rx(param[i],qr[i])
            qc.ry(param[i+1],qr[i])
            qc.rz(param[i+2],qr[i])
        [qc.cx(qr[i],qr[i+1]) for i in range(N-1)]
        qc.barrier()
    [qc.h(qr[i]) for i in range(N)]
    qc.barrier()
    #qc.x(qr[1])
    #qc.x(qr[2])
    #qc.unitary
    [qc.measure(qr[i], cr[i]) for i in range(N)]
    return qc
```

```python
N=3
qc2 = circuit2(N,param,3)
qc2.draw(output="mpl")
```

```python
counts2 = execute(qc2,Aer.get_backend('qasm_simulator'),shots = 100).result().get_counts()
print("counts:",counts2)
counts2 = conv_qubitbook(counts2)
counts2 = dict(sorted(counts2.items()))
print("new_counts:",counts2)
print("qubit list:",qubit_list(N))
plot_histogram(conv_qubitbook(counts2))
```

Now with the circuit2 we can get the counts for the expressions $<\psi|\sigma^{X}\otimes \mathbb{1}\otimes\mathbb{1}|\psi>$, $<\psi|\mathbb{1}\otimes\sigma^{X}\otimes \mathbb{1}|\psi>$ and $<\psi|\mathbb{1}\otimes\mathbb{1}\otimes\sigma^{X}|\psi>$ for the case where the Hadamard gate is at the 1st, 2nd or 3rd register respectivelly.

We have to be able to add the counts from these different measurements in order to compute H2

```python
def compute_H2(N,NUM_SHOTS,param,h,layers):
    a2 = []
    H2 = 0
    qc2 = circuit2(N,param,layers)
    job2 = execute(qc2,Aer.get_backend('qasm_simulator'),shots=NUM_SHOTS)
    result2 = job2.result()
    counts2 = result2.get_counts()
    counts2 = conv_qubitbook(counts2)
    counts2 = dict(sorted(counts2.items())) #sorts the dictionary of counts according to their key
    [a2.append([int(d) for d in str(list(counts2)[i])]) for i in range(len(list(counts2)))]
    #print("counts after sorting:",counts)
    #split dictionary "count" keys into array elements
    #To account for the case where a register gives only one qubit with probability 1 after all shots
    shots2 = sum(counts2.values())
    P2 = [(i, c/shots2) for i, c in counts2.items()]
    #print("register[{}] a2={}".format(register,a2))
    #print("register[{}] P={}".format(register,P2))
    for j in range(N): #runs over the N-1 pairs of consecutive qubits
        for i in range(2**N): #runs over the set of 2**N states
            if a2[i][j]==0:
                H2 += h[j]*P2[i][1]
            if a2[i][j]==1:
                H2 -= h[j]*P2[i][1]
            #print("H2 (i={},j={}):{}".format(i,j,H2))
    #print("H2 =",H2)
    return H2
```

```python
NUM_SHOTS = 1000
N=3
h = couplings(N)[0]
compute_H2(N,NUM_SHOTS,param,h,2)
```

## Lets combine these to compute the expectation value $<\psi|H|\psi>$

```python
def expval_ising(param):
    
    NUM_SHOTS = 10000
    energy = 0
    H1 = 0
    H2 = 0
    a  = []
    a2 = []
    N  = 3
    layers = 1
    h  = couplings(N)[0]
    J  = couplings(N)[1]
    parameters = param 
    H1 = compute_H1(N,NUM_SHOTS,param,J,layers)
    H2 = compute_H2(N,NUM_SHOTS,param,h,layers)
    energy = -H1-H2
    print("H  = -H1-H2 = ",energy)
    return energy
```

# <font color ='#4d4dff'> SPSA 

### <font color ='#ff8000'> 1 Layers:

An appropriate optimizer for optimizing a noisy objective function is the Simultaneous Perturbation Stochastic Approximation optimizer (SPSA). SPSA approximates the gradient of the objective function with only two measurements. It does so by concurrently perturbing all of the parameters in a random fashion, in contrast to gradient decent where each parameter is perturbed independently. When utilizing VQE in either a noisy simulator or on real hardware, SPSA is a recommended as the classical optimizer.

```python
from qiskit import QuantumCircuit, BasicAer, execute
from qiskit.visualization import plot_state_city
import numpy as np
%matplotlib inline
opts = {"initial_statevector":np.array([1,0,0,1]/np.sqrt(2))}
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])
qc.draw(output="mpl")
```

```python
backend = BasicAer.get_backend('statevector_simulator')
job = execute(qc, backend).result()
plot_state_city(job.get_statevector(qc), color=['midnightblue', 'midnightblue'],
     title="New State City")
```
